```python
#Extract data from Matlab files

def get_tensors_from_matlab(matlab_files_name):
  """
  Extracts the acquisitions of each Matlab file in the dictionary matlab_files_name.
  """
  acquisitions = {}
  for key in matlab_files_name:
    file_name = matlab_files_name[key] #key-value형태로 저장해둔 것에서
    matlab_file = scipy.io.loadmat(file_name) #key호출로 value값을 옮길 수 있나보다
    for position in ["DE"]:  #DE의 정체는?
      keys = [key for key in matlab_file if key.endswith(position+"_time")]
      if len(keys) > 0:
        array_key = keys[0] #첫번째 요소에 뭐가 들어갈까?
        acquisitions[key] = matlab_file[array_key].reshape(1, -1)[0]
  return acquisitions
```

```python
#Segment data
def cwru_segmentation(acquisitions, sample_size=512, max_samples=None): #윈도우 사이즈 별로 짤라야하지 않나?
  """
  Segments the acquisitions.
  sample_size is the size of each segment
  max_samples is used for debug purpose and reduces the number of samples from acquisitions
  """
  origin = []
  data = np.empty((0, sample_size, 1)) #make void np array
  n = len(acquisitions)
  for i, key in enumerate(acquisitions):
    acquisition_size = len(acquisitions[key])
    n_samples = acquisition_size // sample_size
    if max_samples is not None and max_samples > 0 and n_samples > max_samples:
      n_samples = max_samples
    print("{}/{} --- {}: {}".format(i+1, n, key, n_samples))
    origin.extend([key for _ in range(n_samples)])
    data = np.concatenate((data,
                           acquisitions[key][:(n_samples* sample_size)].reshape(
                               (n_samples, sample_size, 1))))
  return data, origin
```

```python
#Selecting samples
import re

def select_samples(regex, X, y):
  """
  Selects samples which has some regex pattern in its name.
  """
  mask = [re.search(regex, label) is not None for label in y]
  return X[mask], y[mask]

def join_labels(regex, y):
  """
  Excludes some regex patterns from the labels
  making some samples to have the same label
  """
  groups = list(range(len(y)))
  for i, label in enumerate(y):
    match = re.search(regex, label)
    groups[i] = match.group(0) if match else None
  return groups

def get_groups(regex, y):
  """
  Generates a list of groups of samples with the same regex pattern in its label.
  """
  groups = list(range(len(y)))
  for i, label in enumerate(y):
    match = re.search(regex, label)
    groups[i] = match.group(0) if match else None
  return groups
```

```python
from keras.layers.convolutional import Conv2D
from keras.layers import Dense
from keras.layers.convolutional import MaxPooling2D,Conv1D,MaxPooling1D
from keras.layers.core import Flatten
from keras.layers import BatchNormalization
from keras.layers import Activation
import keras
```

```python
def compiled_CNFN_model():
    model = Sequential()
    model.add(Conv1D(5, 32,padding='same', activation='relu', input_shape=(1024, 1)))#, activation='relu' keras.optimizers.Adam() keras.optimizers.RMSprop()
    model.add(Conv1D(3, 16, activation='relu'))#, activation='relu'
    #model.add(MaxPooling1D(3, data_format='channels_first'))
    model.add(MaxPooling1D(10))
    #model.add(BatchNormalization())
    model.add(Flatten())
    #f_layer = FuzzyLayer(32)
    #model.add(f_layer)
    model.add(BatchNormalization(trainable=True))
    model.add(Dense(16))
    model.add(Activation('softmax'))
    model.compile(
        loss = 'categorical_crossentropy',
        optimizer = tf.keras.optimizers.Adam(), 
        metrics= ['accuracy']
        )
    return model
```

```python
compiled_CNFN_model().summary()
```

```python
model = compiled_CNFN_model()
history = model.fit(X_train, y_train, epochs=120, batch_size=64, verbose=1, validation_split=0.1)
```
